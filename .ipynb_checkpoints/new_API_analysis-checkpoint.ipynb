```python
def find_old_link_from_sites(sites):
    siteA=sites[0]
    siteB=sites[1]
    old_links=old.json()["links"]
    for links in old_links:
        if links['sites'][0]==siteA and links['sites'][1]==siteB:
            return links
    print("Nada")
```

```python
def find_link(link_list, link):
    for links in link_list:
        #print(links)
        if (link == links['id']):
            return links
    print("Nada")
```

```python
def filter_sets(listA, listB): 
    setA=set(listA)
    setB=set(listB)
    
    return list(setA.difference(setB))
```

```python
def unique(list1): 
      
    # insert the list to the set 
    list_set = set(list1) 
    # convert the set to the list 
    unique_list = (list(list_set)) 
    #for x in unique_list: 
        #print(x)
    return len(unique_list)
```
